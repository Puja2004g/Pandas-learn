```python
#count the null values of columns
data.isnull().sum()
```

```python
data.dropna(axis=0)   ## default value of axis = 0  (drop all rows where any column value is null)
data.dropna(how='all')   ## default value of how = 'any'  (drop all rows where all column value is null)
```

```python
data.dropna(axis=1)   ## default value of axis = 0  (drop all columns where any row value is null)
data.dropna(how='any')   ## default value of how = 'any'  (drop all rows where any column value is null)
```

```python
data.dropna(subset=["Age"])  ## (drop all rows where all Age value is null)
```

```python
#print the mean age of the data 
mean_age = round(data['Age'].mean())
mean_age
```

```python
#distinct values entered in specified columns
data['Embarked'].value_counts()
```

```python
#fill the mean age value to the null values of the age column
data['Age'] = data['Age'].fillna(mean_age)
data['Cabin'] = data['Cabin'].fillna("NA")
data['Embarked'] = data['Embarked'].fillna('Q')
```

```python
#sort the data in ascending order
data_sorted = data.sort_values("Name",ascending=True)
data_sorted.head()
```

```python
#sort the data in descending orde
data_sorted = data.sort_values("Name",ascending=False)
data_sorted.head()
```

```python
#sort the data in ascneding order and indexed it
data_sorted = data.sort_values("Name",ascending=False)
data_sorted = data_sorted.reset_index()
data_sorted.head()
```

```python
#drop the specified row 2,3
data_sorted = data_sorted.drop([2,3],axis=0)  ## remove row
data_sorted
```

```python
#drop the specifed column (index)
data_sorted = data_sorted.drop(["index"],axis=1)  ## remove column
data_sorted.head()
```

```python
#export the data to csv
data_sorted.to_csv("mytitanic_data.csv",index=False)   ## export to csv
```

```python
#export the data to excel
data_sorted.to_excel("mytitanic_data.xlsx",index=False)  ## export to excel
data_sorted
```
